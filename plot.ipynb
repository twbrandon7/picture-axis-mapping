```python
from matplotlib import pyplot as plt
from matplotlib.patches import Rectangle


def plot_line(ax, xy1: tuple, xy2: tuple, **kwargs):
    """ Plot line segement on axis
    """
    # plot line
    ax.plot((xy1[0], xy2[0]), (xy1[1], xy2[1]), **kwargs)
    # plot markers at starting and ending points
    ax.plot(*xy1, ms=15, marker='_', mec='g', mew=3, ls="none")
    ax.plot(*xy2, ms=15, marker='_', mec='g', mew=3, ls="none")
```

```python
# The background image
image = plt.imread("./images/photon_emission.png")
# Get axes
ax = plt.gca()
# Plot backgrund image
ax.imshow(image)

# Add the rectangles to the current axes
ax.add_patch(Rectangle(xy=(295, 198), width=215, height=120, label="A", fill=False, edgecolor='red', linewidth=2))
ax.add_patch(Rectangle(xy=(295, 234), width=100, height=35, label="B", fill=False, edgecolor='blue', linewidth=2))
ax.add_patch(Rectangle(xy=(295, 35), width=87, height=70, label="C", fill=False, edgecolor='orange', linewidth=2))

# Plot lines
plot_line(ax, (350, 35 + 70), (350, 198), linestyle='-', color='green', linewidth=2)
ax.text(355, 160, "distance", color="red", size=16)

plt.legend()
plt.show()
```
